## Ejercico 1: 

```python
ys, ts, h = ODE_euler(a=0,b=2.0,f= lambda t,y: y-t**2+1 ,y_t0=0.5,N=10)
ys5, ts5, h5 = ODE_euler(a=0,b=2.0,f= lambda t,y: y-t**2+1 ,y_t0=0.5,N=5)
ys20, ts20, h20 = ODE_euler(a=0,b=2.0,f= lambda t,y: y-t**2+1 ,y_t0=0.5,N=20)
```

## Ejercicio 2: 


```python
ys, ts, h = ODE_euler(a=1.0,b=2.0,f= lambda t,y: y/t-(y/t)**2 ,y_t0=1,N=10)
ys5, ts5, h5 = ODE_euler(a=1.0,b=2.0,f= lambda t,y: y/t-(y/t)**2 ,y_t0=1,N=5)
ys20, ts20, h20 = ODE_euler(a=1.0,b=2.0,f= lambda t,y: y/t-(y/t)**2 ,y_t0=1,N=20)
```

```python
def f2_real(t):
    return t/(1+np.log(t)) 

ys2_real = [f2_real(t) for t in ts]

plt.figure(figsize=(10,6))
plt.plot(ts, ys2_real, 'o-',label='Funcion real')
plt.plot(ts5, ys5 ,linestyle='dashed',label='Funcion aproximada N=5 ')
plt.plot(ts, ys ,linestyle='dashed',label='Funcion aproximada N=10 ')
plt.plot(ts20, ys20 ,linestyle='dashed',label='Funcion aproximada N=20 ')
plt.title('Funcion real vs Metodo de Euler')
plt.xlabel('t')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```
